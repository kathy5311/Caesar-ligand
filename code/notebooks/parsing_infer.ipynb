```python
# 각 카테고리별 label 값을 추출
elem_list_label = []
for sub_list in label_list3:
    try:
        elem_list_label.append(sub_list['elem'].index(1))
    except ValueError:
        continue  
func_list_label=[]
for sub_list in label_list3:
    try:
        func_list_label.append(sub_list['func'].index(1))
    except ValueError:
        continue 

numH_list_label = []
for sub_list in label_list3:
    try:
        numH_list_label.append(sub_list['numH'].index(1))
    except ValueError:
        continue 


aromatic_labels2 = [int(item["aromatic"]) for item in label_list3]
numCH3_labels2 = [int(item["numCH3"]) for item in label_list3]
ring_labels2 = [int(item["ring"]) for item in label_list3]


hybrid_list_label=[]
for sub_list in label_list3:
    try:
        hybrid_list_label.append(sub_list['hybrid'].index(1))
    except ValueError:
        continue 

# 히스토그램 또는 바 차트를 사용하여 분포 시각화
def plot_distribution(label_data, title, xlabel):
    plt.figure(figsize=(10, 6))
    sns.histplot(label_data, kde=False, bins=np.arange(0, max(label_data) + 1.5) - 0.5, color="blue")
    plt.xticks(range(max(label_data)+1))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Count')
    plt.show()


# 각 카테고리 분포 시각화
plot_distribution(elem_list_label, "Element Label Distribution", "Sum of Element Labels")
plot_distribution(func_list_label, "Function Label Distribution", "Sum of Function Labels")
plot_distribution(numH_list_label, "Number of Hydrogens Label Distribution", "Number of Hydrogens")
plot_distribution(aromatic_labels2, "Aromatic Label Distribution", "Aromatic")
plot_distribution(numCH3_labels2, "Number of CH3 Label Distribution", "Number of CH3")
plot_distribution(ring_labels2, "Ring Label Distribution", "Ring")
plot_distribution(hybrid_list_label, "Hybridization Label Distribution", "Hybridization")
```

```python
def plot_comparison_distribution(label_data, pred_data, title, xlabel):
    plt.figure(figsize=(10, 6))
    sns.histplot(label_data, kde=False, bins=np.arange(0, max(label_data + pred_data) + 1.5) - 0.5, color="red", label='Label', alpha=0.5)
    sns.histplot(pred_data, kde=False, bins=np.arange(0, max(label_data + pred_data) + 1.5) - 0.5, color="blue", label='Prediction', alpha=0.5)
    plt.xticks(range(max(label_data + pred_data) + 1))
    plt.title(title)
    plt.xlabel(xlabel+'(index)')
    plt.ylabel('Count')
    plt.legend()
    plt.show()

# 각 카테고리 분포 시각화
plot_comparison_distribution(elem_list_label, elem_list_pred, "Element Comparison Distribution", "Element")
plot_comparison_distribution(func_list_label, func_list_pred, "Function Comparison Distribution", "Function")
plot_comparison_distribution(numH_list_label, numH_list_pred, "Number of Hydrogens Comparison", "Number of Hydrogens")
plot_comparison_distribution(aromatic_labels2, aromatic_pred, "Aromatic Comparison", "Aromatic")
plot_comparison_distribution(numCH3_labels2, numCH3_pred, "Number of CH3 Comparison", "Number of CH3")
plot_comparison_distribution(ring_labels2, ring_pred, "Ring Comparison", "Ring")
plot_comparison_distribution(hybrid_list_label, hybrid_list_pred, "Hybridization Comparison", "Hybridization")
```
