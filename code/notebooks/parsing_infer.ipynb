```python
def plot_comparison_distribution(label_data, pred_data, title, xlabel):
    plt.figure(figsize=(10, 6))
    sns.histplot(label_data, kde=False, bins=np.arange(0, max(label_data + pred_data) + 1.5) - 0.5, color="red", label='Label', alpha=0.5)
    sns.histplot(pred_data, kde=False, bins=np.arange(0, max(label_data + pred_data) + 1.5) - 0.5, color="blue", label='Prediction', alpha=0.5)
    plt.xticks(range(max(label_data + pred_data) + 1))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Count')
    plt.legend()
    plt.show()

# 각 카테고리 분포 시각화
plot_comparison_distribution(elem_list_label, elem_list_pred, "Element Comparison Distribution", "Element")
plot_comparison_distribution(func_list_label, func_list_pred, "Function Comparison Distribution", "Function")
plot_comparison_distribution(numH_labels2, numH_pred, "Number of Hydrogens Comparison", "Number of Hydrogens")
plot_comparison_distribution(aromatic_labels2, aromatic_pred, "Aromatic Comparison", "Aromatic")
plot_comparison_distribution(numCH3_labels2, numCH3_pred, "Number of CH3 Comparison", "Number of CH3")
plot_comparison_distribution(ring_labels2, ring_pred, "Ring Comparison", "Ring")
plot_comparison_distribution(hybrid_labels2, hybrid_pred, "Hybridization Comparison", "Hybridization")
```
